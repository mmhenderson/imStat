```python
import sys
import os
import struct
import time
import copy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
%matplotlib inline

fpX = np.float32

import torch 
root = os.path.dirname(os.path.dirname(os.getcwd()))
sys.path.append(os.path.join(root, 'code'))
from plotting import plot_utils, summary_plots, load_fits
plot_utils.set_plotting_defaults()
from utils import roi_utils, nsd_utils, default_paths
from feature_extraction import linear_discr
from utils import stats_utils
```

```python
# Load results of the fitting procedure

subject = 1
volume_space=True
fitting_type = 'sketch_tokens'

root = os.path.dirname(os.path.dirname(os.getcwd()))
out, fig_save_folder = load_fits.load_fit_results(subject, volume_space, fitting_type, \
                                                  n_from_end=4, root=root, verbose=True)
load_fits.print_output_summary(out)

if not os.path.exists(fig_save_folder):
    os.makedirs(fig_save_folder)
```

```python
roi_def = roi_utils.get_combined_rois(subject,include_all=True,include_body=True,volume_space=True,verbose=False) 
skip_inds = [10,11,15,23]
retlabs, facelabs, placelabs, bodylabs, \
        ret_names, face_names, place_names, body_names = roi_def
roi_names = ret_names+face_names+place_names+body_names
```

```python
recon_corrcoef, sem_proj_slope, sem_proj_inter, sem_proj_r2, \
                    recon_sem_proj, actual_sem_proj, axes_to_do = out['voxel_recs']
recon_corrcoef = np.nan_to_num(recon_corrcoef)
pop_recons, pop_recon_r2, pop_recon_corrcoef, pop_recon_angle = out['pop_recs']
pop_recon_corrcoef = np.nan_to_num(pop_recon_corrcoef)
```

```python
# load semantic labels for trials and pre-computed LDA axes

lda_result_list = []
sketch_token_feat_path = default_paths.sketch_token_feat_path
for aa, discrim_type in enumerate(axes_to_do):
    features_file = os.path.join(sketch_token_feat_path, 'LDA', \
                                              'S%d_LDA_%s.npy'%(subject, discrim_type))
    print('loading from %s'%features_file)
    lda_result = np.load(features_file, allow_pickle=True).item()
    lda_result_list.append(lda_result)
    
```

```python
# Load information about images for each trial

sessions = np.arange(out['up_to_sess'])
image_order = nsd_utils.get_master_image_order()
session_inds = nsd_utils.get_session_inds_full()
sessions = np.array(sessions)
inds2use = np.isin(session_inds, sessions)
image_order = image_order[inds2use]
shared_1000_inds = image_order<1000
image_order_val = image_order[shared_1000_inds]
image_order_trn = image_order[~shared_1000_inds]
```

```python
# Based on recons in sketch tokens space, get cross-validated decodability of each 
# semantic dimension.

n_rois = pop_recons.shape[2]
n_cv = 10;
n_trials = len(image_order_val)
n_trials_per_cv = int(np.ceil(n_trials/n_cv))
cv_inds = np.tile(np.arange(n_cv), [1,n_trials_per_cv])[0,0:n_trials]

dprime_vals = np.zeros((n_rois,len(axes_to_do)))

for aa, discrim_type in enumerate(axes_to_do):
    
    labels = np.any(np.array(lda_result_list[aa]['labels_actual']), axis=0)
    labels = labels[image_order_val]
    
    for rr in range(n_rois):
        
        if ~np.all(pop_recons[:,:,rr]==0):
            
            pred_labels = np.zeros((n_trials,))
            for cv in range(n_cv):
                trninds = cv_inds!=cv
                tstinds = cv_inds==cv
                scores, scalings, trn_acc, trn_dprime, clf = \
                    linear_discr.do_lda(values = pop_recons[trninds,:,rr], categ_labels = labels[trninds], \
                        verbose=False, balance_downsample=False, rndseed=None)
                pred_labels[tstinds] = clf.predict(pop_recons[tstinds,:,rr])
            tst_dprime = stats_utils.get_dprime(pred_labels, labels, np.unique(labels))
            dprime_vals[rr,aa] = tst_dprime
            
        else:
            
            dprime_vals[rr,aa] = np.nan
```

```python
for aa, discrim_type in enumerate(axes_to_do):
    plt.figure();
    inds2use = ~np.isin(np.arange(n_rois), skip_inds)
#     inds2use = ~np.isnan(dprime_vals[:,aa])
    n_rois_plot = np.sum(inds2use)
    plt.bar(np.arange(n_rois_plot),dprime_vals[inds2use,aa])
    plt.xticks(np.arange(n_rois_plot), labels=np.array(roi_names)[inds2use], rotation=90);
    plt.ylabel('d-prime')
    plt.title('discriminability of %s'%discrim_type)
    plt.ylim([0,2])
```
